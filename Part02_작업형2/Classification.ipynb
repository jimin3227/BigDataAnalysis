```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
# 1. 데이터 불러오기
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# 2. 결측치 처리
train.isnull().sum()
```

```python
x_train = train.iloc[:,:15]
y_train = train.iloc[:, -1]

map = {'<=50K':0, '>50K':1}
y_train = y_train.replace(map)

x_full = pd.concat([x_train, test], axis=0)
x_full.info()
```

```python
x_full.isnull().sum()
```

```python
## id 컬럼 제거
x_full.drop(columns=['id'], inplace=True)
```

```python
x_full['age'].fillna(x_full['age'].median(), inplace=True)
x_full['workclass'].fillna(x_full['workclass'].mode()[0], inplace=True)
x_full['occupation'].fillna(x_full['occupation'].mode()[0], inplace=True)
x_full['hours.per.week'].fillna(x_full['hours.per.week'].mean(), inplace=True)
x_full['native.country'].fillna(x_full['native.country'].mode()[0], inplace=True)
x_full.isnull().sum()
```

```python
x_full = pd.get_dummies(x_full, dtype=int, drop_first=True)
```

```python
x_train = x_full.iloc[:29304,:]
x_test = x_full.iloc[29304:, :]

x_train.shape, x_test.shape
```

```python
#3. 스케일링
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
#4. train_test_split 나누기
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.3, random_state=0)
```

```python
#5. 랜덤 포레스트 Classifier
rfc = RandomForestClassifier()
rfc_model = rfc.fit(x_train, y_train)
y_pred_proba = rfc_model.predict_proba(x_val)
y_pred = rfc_model.predict(x_val)
```

```python
#6. 평가하기
##1) roc_auc
roc_auc = roc_auc_score(y_val, y_pred_proba[:,1])
print("ROC_AUC: ", roc_auc)

##2) accuracy
accuracy = accuracy_score(y_val, y_pred)
print("Accuracy: ", accuracy)

##3) f1-score
f1score = f1_score(y_val, y_pred)
print("F1-score: ", f1score)
```

```python
y_pred = rfc.predict_proba(x_test)
submit = pd.DataFrame({'pred': y_pred[:,1]})
submit.to_csv("result.csv", index=False)
```
