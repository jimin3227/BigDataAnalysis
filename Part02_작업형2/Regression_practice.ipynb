### 03. 중고차 가격 예측

```python
import pandas as pd
```

```python
#1.
train = pd.read_csv('car_train.csv')
test = pd.read_csv('car_test.csv')
```

```python
train.info()
```

```python
x_train = train.iloc[:, train.columns != 'Price']
y_train = train['Price']
x_test = test

x_train.shape, test.shape
```

```python
x_full = pd.concat([x_train, x_test])
x_full.shape
```

```python
cols = list(x_full.columns[x_full.dtypes == object])
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in cols:
    x_full[col] = le.fit_transform(x_full[col])

x_train = x_full.iloc[:6732,:]
x_test = x_full.iloc[6732:,:]
```

```python
x_train.info()
```

```python
x_train.head()
```

```python
x_test
```

```python
#3. scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
#4.
from sklearn.model_selection import train_test_split
x_tr, x_val, y_tr, y_val = train_test_split(x_train, y_train, test_size=0.3, random_state=0)
```

```python
#5.
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr_model = rfr.fit(x_tr, y_tr)
y_pred = rfr_model.predict(x_val)
```

```python
from sklearn.metrics import root_mean_squared_log_error
rmsle = root_mean_squared_log_error(y_val, y_pred)
rmsle
```

```python
pred = rfr_model.predict(x_test)
submit = pd.DataFrame({'pred': pred})
submit.to_csv('result6.csv', index=False)
```
