```python
import pandas as pd
import numpy as np

df=pd.DataFrame({
    "Menu": ['Americano', 'Latte', 'Espresso', 'Caffe Mocha', 'Vanilla Latte'],
    "Price": [4500, 5000, 4000, 5900, 5300],
    "Calories": [10, 110, np.nan, 210, np.nan],
    "Beans": ["Guatemala", "Brazil", "Guatemala", np.nan, np.nan]
})
df
```

```python
#Q1.
df['Calories'].fillna(df['Calories'].min(), inplace=True)
df.head()
```

```python
#Q2.
df['Beans']= df['Beans'].fillna(df['Beans'].mode()[0])
df
```

```python
#Q3.
df[df['Price']>=5000].shape[0]
```

```python
#Q4.
df["Event"]= df['Price']*0.5
df
```

```python
df['Event'] = df['Event'].astype(int)
df
```

```python
#Q5.
df.drop('Calories', axis=1, inplace=True)
df
```
