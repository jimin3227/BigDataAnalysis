```python
import pandas as pd
import numpy as np

df=pd.DataFrame({
    "Menu": ['Americano', 'Latte', 'Espresso', 'Caffe Mocha', 'Vanilla Latte'],
    "Price": [4500, 5000, 4000, 5900, 5300],
    "Calories": [10, 110, np.NaN, 210, np.NaN],
    "Beans": ["Guatemala", "Brazil", "Guatemala", np.NaN, np.NaN]
})
df
```

```python
#Q1.
min = df['Calories'].min()
df['Calories']=df["Calories"].fillna(min)
df
```

```python
#Q2.
mode = df['Beans'].mode()[0]
df["Beans"]=df["Beans"].fillna(mode)
df
```

```python
#Q3.
cond = df['Price'] >= 5000
sum(cond)
```

```python
#Q4.
df["Event"] = df["Price"]*0.5
df
```

```python
#Q5.
df.drop(columns=["Calories"], inplace=True)
df
```
